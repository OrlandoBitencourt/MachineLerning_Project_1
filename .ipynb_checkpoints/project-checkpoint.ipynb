```python
# importing libs

import os
from io import open
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
import seaborn as sns


pd.options.display.max_columns = 150
```

```python
# set up the data

df = pd.read_csv('dataset\data-final.csv', sep='\t')
```

```python
df.head()
```

```python
# looking the database info

pd.options.display.float_format = "{:.2f}".format
df.describe()
```

```python
# remove other columns (keeping only the questions responses)

df.drop(df.columns[50:], axis=1, inplace=True)
```

```python
# verify the records in EXT1 column (3720 with value equals zero)

df["EXT1"].value_counts()
```

```python
# lets see all the value 0 itemns in our dataframe

df[(df == 0.00).all(axis=1)].describe()
```

```python
# cleaning up the dataframe

df = df[(df > 0.00).all(axis=1)]
```

```python
# checking it again

df["EXT1"].value_counts()
```

```python
# importing libs

from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
# instanciating kmeans and visualizer

kmeans = KMeans()
visualizer = KElbowVisualizer(kmeans, k=(2,10))
```

```python
# getting a random piece of the data with 6000 observations

data_sample = df.sample(n=6000, random_state=1)
```

```python
# lets see the data

visualizer.fit(data_sample)
visualizer.poof()
```

```python
# grouping people in five groups

kmeans = KMeans(n_clusters=5)
k_fit = kmeans.fit(df)
```

```python
predictions = k_fit.labels_
df['Clusters'] = predictions
```

```python
df.head()
```

```python
# checking clusters

df["Clusters"].value_counts()
```

```python
# grouping data by cluster

df.groupby('Clusters').mean()
```

```python
# selecting the columns with the range of the questions

col_list = list(df)
ext = col_list[0:10]
est = col_list[10:20]
agr = col_list[20:30]
csn = col_list[30:40]
opn = col_list[40:50]
```

```python
# creating a new dataframe to calculating the avarage of each individual group to identify if theres any pattern

data_add = pd.DataFrame()
data_add['extroversion'] = df[ext].sum(axis=1)/10
data_add['neurotic'] = df[est].sum(axis=1)/10
data_add['agreeable'] = df[agr].sum(axis=1)/10
data_add['conscientious'] = df[csn].sum(axis=1)/10
data_add['open'] = df[opn].sum(axis=1)/10
data_add['clusters'] = predictions
```

```python
# showing up the avarage by group

data_add.groupby('clusters').mean()
```

```python
# visualizing the data

data_by_clusters = data_add.groupby('clusters').mean()
```

```python
plt.figure(figsize=(25,5))
for i in range(0,5):
    plt.subplot(1,5, i+1)
    plt.bar(data_by_clusters.columns, data_by_clusters.iloc[:, i], color='blue', alpha=0.2)
    plt.plot(data_by_clusters.columns, data_by_clusters.iloc[:, i], color='black')
    plt.title('Group' + str(i))
    plt.xticks(rotation=45)
    plt.ylim(0,4)
```
